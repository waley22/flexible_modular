```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
demand_arr = np.array([237.56055795, 210.48812593, 363.62573597, 374.55363561])

plt.figure(figsize = (10, 4))
plt.subplot(121)
plt.title("(a)")
plt.bar([1, 2, 3, 4], demand_arr, color = "#8DA7D9", width = 0.55, label = "Demand")
plt.xlim([0.5, 4.5])
plt.ylim([0, 400])
plt.xticks([1, 2, 3, 4])
plt.xlabel("Customer")
plt.ylabel("Demand at t = 1")
plt.vlines([1, 2, 3, 4], demand_arr, [demand_arr.mean() for i in range(4)], color = "red")
plt.plot([0, 5], [demand_arr.mean(), demand_arr.mean()], color = "g", ls = "--", label = "Mean demand")
plt.text(0.6, demand_arr.mean() + 10, "Mean demand in t = 1", color = "g")

# plt.savefig("./illus_FI1.png", transparent = True)

demand_arr_2 = np.array([115.06040089, 471.33030979, 221.08467064, 343.40837099]) - 100

plt.subplot(122)
plt.title("(b)")
plt.plot([1, 2, 3, 4], demand_arr_2, color = "#8DA7D9", linewidth = 1.5)
plt.scatter([1, 2, 3, 4], demand_arr_2, color = "#8DA7D9")
plt.xlim([0.5, 4.5])
plt.ylim([0, 400])
plt.xticks([1, 2, 3, 4])
plt.xlabel("t")
plt.ylabel("Demand for customer #1")

for i in range(1, 4):
    plt.plot([i, i + 1], [demand_arr_2[i - 1], demand_arr_2[i - 1]], color = "k", ls = "--")
    plt.plot([i + 1, i + 1], [demand_arr_2[i - 1], demand_arr_2[i]], color = "r")

plt.savefig("./illus_FIs.png", transparent = False)

print(demand_arr, demand_arr_2)
```
